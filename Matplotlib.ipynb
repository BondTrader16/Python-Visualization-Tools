```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Chart Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')

plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

```python
fig=plt.figure()

axes=fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y,'b')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Chart Title')
plt.show()
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'b')
axes1.set_xlabel('X1 Axis')
axes1.set_ylabel('Y1 Axis')
axes1.set_title('Chart Title1')

axes2.plot(y,x,'r')
axes2.set_xlabel('X2 Axis')
axes2.set_ylabel('Y2 Axis')
axes2.set_title('Chart Title2')
plt.show()
```

```python
fig, axes = plt.subplots()

axes.plot(x,y,'r')
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('Chart Title')
```

```python
fig, axes=plt.subplots(nrows=1,ncols=2)
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X Square')
ax.plot(x,x**3,label='X Cube')
ax.legend(loc=0)
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='Purple',lw=2,ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
